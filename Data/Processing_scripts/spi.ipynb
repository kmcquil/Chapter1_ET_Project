```python
import ee
import geemap 
import os
ee.Initialize()
```

```python
roi = ee.FeatureCollection('users/kmcquil/SBR_XL_simplified').geometry()
spi = ee.ImageCollection("GRIDMET/DROUGHT")
spi = spi.filterDate('1984-01-01', '2020-12-31').filterBounds(roi)
```

```python
# Define time range
# create list for years
years = range(1984, 2021)
 
# make a list with months
months = range(1,13)
 

# calculate the monthly mean
def calcMonthlyMean(imageCollection):
    mylist = ee.List([])
    for y in years:
        for m in months:
            w = imageCollection.filter(ee.Filter.calendarRange(y, y, 'year')).filter(ee.Filter.calendarRange(m, m, 'month')).mean()
            mylist = mylist.add(w.set('year', y).set('month', m).set('date', ee.Date.fromYMD(y,m,1)).set('system:time_start',ee.Date.fromYMD(y,m,1)).set('system:index',  ee.String(ee.Number(y).format('%04d')).cat(ee.Number(m).format('%02d'))))
    return ee.ImageCollection.fromImages(mylist)
```

```python
spi90 = spi.select('spi90d') # subset to the spi time period of interest 
mo_spi90 = ee.ImageCollection(calcMonthlyMean(spi90)) # run the calcMonthlyMean function

# export disalexi data to a google drive folder for further analysis 
geemap.ee_export_image_collection_to_drive(ee_object=mo_spi90, 
folder='SPI90_OG', 
region=roi,
scale = 4000
)
```

```python
# now grab spi 12 month just for comparison sake 
spi1y = spi.select('spi1y') # subset to the spi time period of interest 
mo_spi1y = ee.ImageCollection(calcMonthlyMean(spi1y)) # run the calcMonthlyMean function

# export disalexi data to a google drive folder for further analysis 
geemap.ee_export_image_collection_to_drive(ee_object=mo_spi1y, 
folder='SPI1Y_OG', 
region=roi,
scale = 4000
)
```
